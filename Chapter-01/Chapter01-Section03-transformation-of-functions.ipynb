```python
import numpy as np
import matplotlib.pyplot as plt
```

- Set the Domain: the x values from -a to -a with npoint values

```python
a = 5
npoints = 200
x = np.linspace(-a, a, npoints)
```

```python
#print(x) #Remove the first hashtag to see the values of x.
```

- Define the functions: example below is for power functions
- Do the same for the vertical, horizontal, stretching and reflection as shown on page 37.

```python
c = 3
n = 2

y0 = x**(n)
y1 = x**(n) + c
y2 = x**(n) - c
y3 = (x + c)**(n) 
y4 = (x - c)**(n)
```

```python
#print(y1) #Remove the first hashtag to see the values of y1.
#print(y4) #Remove the first hashtag to see the values of y2.
```

- Plot the functions on the domain [-a,a]
- Add a title, labels, grid, labels, and legend

```python
plt.plot(x, y0, label=r'$x^n$', color='black')
plt.plot(x, y1, label=r'$x^n + c$', color='red')
plt.plot(x, y2, label=r'$x^n - c$', color='green')
plt.plot(x, y3, label=r'$(x + c)^n$', color='blue')
plt.plot(x, y4, label=r'$(x - c)^n$', color='purple')
plt.title('Vertical/Horizontal Shifting')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
