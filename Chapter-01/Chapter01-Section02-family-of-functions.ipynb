### Chapter 1, Section 2: Family of Functions
- This section introduces you to some common functions (see page 32). We will use Jupyter to plot these functions.
- The example below is for reciprocal functions. Repeat this process for the other families of functions listed on page 32 (e.g., linear, root, exponential, logarithmic, and trigonometric).
- Plot at least five graphs for each function type, and feel free to be creative.
- When analyzing the behavior of functions, remember that the domain is important.

```python
import numpy as np
import matplotlib.pyplot as plt
```

- Set the Domain: the x values from -a to -a with n values

```python
a = 1
npoints = 200
x = np.linspace(-a, a, npoints)
```

```python
#print(x) #Remove the first hashtag to see the values of x.
```

```python
y0 = x**(-0)
y1 = x**(-1)
y2 = x**(-2)
y3 = x**(-3)
y4 = x**(-4)
```

```python
#print(y1) #Remove the first hashtag to see the values of y1.
#print(y4) #Remove the first hashtag to see the values of y2.
```

- Plot the functions on the domain [-a,a]
- Add a title, labels, grid, labels, and legend

```python
plt.plot(x, y0, label=r'$x^-0$', color='black')
plt.plot(x, y1, label=r'$x^-1$', color='black')
plt.plot(x, y2, label=r'$x^-2$', color='black')
plt.plot(x, y3, label=r'$x^-3$', color='black')
plt.plot(x, y4, label=r'$x^-4$', color='black')
plt.title('Reciprocal Functions')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
