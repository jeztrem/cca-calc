```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_vals = np.linspace(0, 1, 400)
y_vals = 1 - x_vals**2

y_vals_inv = np.linspace(0, 1, 400)
x_vals_inv = np.sqrt(1 - y_vals_inv)

plt.plot(x_vals, y_vals, label='f(x) = 1 - x²', color='red')
plt.plot(y_vals_inv, x_vals_inv, label='Inverse: f⁻¹(y)', color='green')
plt.plot(x_vals, x_vals, color='black', linestyle=':', label='y = x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and Its Inverse')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
