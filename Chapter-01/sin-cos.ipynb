```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of values from 0 to 2Ï€
x = np.linspace(0, 2 * np.pi, 100)

# Calculate sine and cosine values
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the functions
plt.plot(x, y_sin, label='sin(x)', color='blue')
plt.plot(x, y_cos, label='cos(x)', color='red')
plt.title('Sine and Cosine Functions')
plt.xlabel('x (radians)')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Create an array of x values from 0 to 2Ï€
x = np.linspace(0, 2 * np.pi, 100)

# Compute cos(x) and its derivative -sin(x)
y_cos = np.cos(x)
y_derivative = -np.sin(x)

# Plot both functions
plt.plot(x, y_cos, label='cos(x)', color='red')
plt.plot(x, y_derivative, label='-sin(x)', color='green')
plt.title('Cos(x) and Its Derivative')
plt.xlabel('x (radians)')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define x values
x = np.linspace(0, 2 * np.pi, 400)
y = np.cos(x)

# Points where we'll draw tangent lines
x_points = np.linspace(0, 2 * np.pi, 10)
y_points = np.cos(x_points)
slopes = -np.sin(x_points)  # Derivative of cos(x)

# Set up plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='cos(x)', color='blue')

# Add tangent lines
for x0, y0, m in zip(x_points, y_points, slopes):
    # Define tangent line: y = m(x - x0) + y0
    x_tangent = np.linspace(x0 - 0.3, x0 + 0.3, 10)
    y_tangent = m * (x_tangent - x0) + y0
    plt.plot(x_tangent, y_tangent, color='black', linestyle='--', linewidth=1)

# Style the plot
plt.title('Cos(x) with Tangent Lines')
plt.xlabel('x (radians)')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define x values
x = np.linspace(0, 2 * np.pi, 400)
y = np.cos(x)

# Points where we'll draw tangent lines
x_points = np.linspace(0, 2 * np.pi, 10)
y_points = np.cos(x_points)
slopes = -np.sin(x_points)  # Derivatives (slope of tangent)

# Set up subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Plot cos(x) and tangent lines
axs[0].plot(x, y, label='cos(x)', color='blue')
for x0, y0, m in zip(x_points, y_points, slopes):
    x_tangent = np.linspace(x0 - 0.3, x0 + 0.3, 10)
    y_tangent = m * (x_tangent - x0) + y0
    axs[0].plot(x_tangent, y_tangent, color='orange', linestyle='--', linewidth=1)
axs[0].set_title('Cos(x) with Tangents')
axs[0].set_xlabel('x (radians)')
axs[0].set_ylabel('y')
axs[0].grid(True)
axs[0].legend()

# Plot slope values
axs[1].plot(x_points, slopes, 'go-', label='Slope = -sin(x)', markersize=6)
axs[1].set_title('Slope of Tangents to Cos(x)')
axs[1].set_xlabel('x (radians)')
axs[1].set_ylabel('Slope')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```
