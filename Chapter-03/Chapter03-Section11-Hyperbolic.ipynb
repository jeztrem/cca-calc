```python
# Plotting parameteric equations: circle
import numpy as np
import matplotlib.pyplot as plt

# Define grid
x = np.linspace(-1, 1, 400)
y = np.linspace(-1, 1, 400)
X, Y = np.meshgrid(x, y)

# Radius
r = 0.75

# Implicit equation of a circle
F = X**2 + Y**2 - r**2
# Solving for dy/dx at y=x in quadrant 1
# Then the equation of the tangent is y=x+a
x0 = np.linspace(-0.5, 1, 400)
tangent = -x0 + np.sqrt(2)*r
# Plot contour where F = 0
plt.contour(X, Y, F, levels=[0], colors='red')
plt.plot(x0, tangent, label=r'${tangent}$', color='blue')
plt.title('Circle')
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
# Plotting parameteric equations: cardioid
# You may want to research the cardioid microphone (like at church)
# Define grid
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)

# Parameter is 1 in problem 33
a = 1.0

# Implicit equation of cardioid
F = (X**2 + Y**2 - a*X)**2 - a**2 * (X**2 + Y**2)

# Solving for dy/dx at (0,a) gives a value of 1
# Then the equation of the tangent is y=x+a
x0 = np.linspace(-1, 1, 400)
tangent = x0 + a

# Plot contour where F = 0
plt.contour(X, Y, F, levels=[0], colors='red')
plt.plot(x0, tangent, label=r'${tangent}$', color='blue')
plt.title('Cardioid')
plt.axis('equal')
plt.grid(True)
plt.show()
```
