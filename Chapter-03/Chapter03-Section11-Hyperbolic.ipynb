```python
# Plotting hyperbolic equations
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 397)
sinh = (np.exp(x) - np.exp(-x)) / 2
cosh = (np.exp(x) + np.exp(-x)) / 2
tanh = sinh / cosh
plt.plot(x, sinh, label=r'$sinh(x)$', color='blue')
plt.plot(x, cosh, label=r'$cosh(x)$', color='red')
plt.plot(x, tanh, label=r'$tanh(x)$', color='green')
plt.title('Hyperbolic Functions')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plotting hyperbolic equations
import numpy as np
import matplotlib.pyplot as plt

x_pos = np.linspace(0.1, 10, 100)
x_neg = np.linspace(-10, -0.1, 100)
sinh_neg = (np.exp(x_neg) - np.exp(-x_neg)) / 2
cosh_neg = (np.exp(x_neg) + np.exp(-x_neg)) / 2
csch_neg = 1 / sinh_neg
sech_neg = 1 / cosh_neg
coth_neg = cosh_neg / sinh_neg
plt.plot(x_neg, csch_neg, label=r'$csch(x)$', color='blue')
plt.plot(x_neg, sech_neg, label=r'$sech(x)$', color='red')
plt.plot(x_neg, coth_neg, label=r'$coth(x)$', color='green')
sinh_pos = (np.exp(x_pos) - np.exp(-x_pos)) / 2
cosh_pos = (np.exp(x_pos) + np.exp(-x_pos)) / 2
csch_pos = 1 / sinh_pos
sech_pos = 1 / cosh_pos
coth_pos = cosh_pos / sinh_pos
plt.plot(x_pos, csch_pos,  color='blue')
plt.plot(x_pos, sech_pos,  color='red')
plt.plot(x_pos, coth_pos,  color='green')
plt.title('Hyperbolic Functions')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plotting parameteric equations: hyperbola 
import numpy as np
import matplotlib.pyplot as plt

# Define grid
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)

# parameters of the hyperbola
t = 2

# Implicit equation of a circle
F = X**2 - Y**2 - t**2
plt.contour(X, Y, F, levels=[0], colors='blue')
plt.title('Hyperbola $x^2 - y^2 = 1$')
plt.axis('equal')
plt.grid(True)
plt.show()
```
