```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#graph of example 3 on page 177.
x0 = np.linspace(0, 2, 400)
fx = (x0**1.5)
x1 = np.linspace(-1, 2, 400)
tangent = (3/2)*x1 - 1/2
normal = -(2/3)*x1 + 5/3

# Plot 
plt.plot(x0, fx, label=r'${x^{\frac{3}{2}}}$', color='black')
plt.plot(x1, tangent, label=r'${tangent}$', color='red')
plt.plot(x1, normal, label=r'${normal}$', color='green')
plt.title(' ')
plt.xlabel('x')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#problem 41 page 182
x0 = np.linspace(-1, 1, 400)
fx = (x0**4 + 2 * np.exp(x0))
x1 = np.linspace(-2, 2, 400)
tangent = 2*x1 + 2
normal = -(1/2)*x1 + 2

# Plot 
plt.plot(x0, fx, label=r'${x^4 + 2e^x}$', color='black')
plt.plot(x1, tangent, label=r'${tangent}$', color='red')
plt.plot(x1, normal, label=r'${normal}$', color='green')
plt.title(' ')
plt.xlabel('x')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
