```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#graph of problem 35 on page 189.
x0 = np.linspace(0, 4, 400)
fx = (x0**2)/(1+x0)
x1 = np.linspace(-1, 3, 400)
tangent = (3/4)*x1 - 1/4
normal = -(4/3)*x1 + 11/6

# Plot 
plt.plot(x0, fx, label=r'${\frac{x^2}{1 + x}}$', color='black')
plt.plot(x1, tangent, label=r'${tangent}$', color='red')
plt.plot(x1, normal, label=r'${normal}$', color='green')
plt.title(' ')
plt.xlabel('x')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
