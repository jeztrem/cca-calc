```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

sym.init_printing()
x, y = sym.symbols('x y')
```

```python

f = 3*x**3 + 2*x + 1
#f = 4*x**3 + 3*x**2 + 2*x + 1
eq = sym.Eq(y, f)
inverse_solved = sym.solve(eq, x)
print ("equation:")
display(eq)
print ("inverse:")
display(inverse_solved)
```

```python

inverse_solved = sym.solve(eq, x)[2]
```

```python
f_lambdified = sym.lambdify(x, f, 'numpy')
inv_lambdified = sym.lambdify(y, inverse_solved, 'numpy')
```

```python
x_vals = np.linspace(-1, 1, 400)
y_vals = f_lambdified(x_vals)
inv_vals = inv_lambdified(y_vals)
```

```python


plt.plot(x_vals, y_vals, label='$f(x)$', color='blue')
plt.plot(y_vals, inv_vals, label='$f^{-1}(x)$', color='red')

# find the min and max for the symmetry line and plot it
sym_min = min(np.min(x_vals), np.min(y_vals), np.min(inv_vals))
sym_max = max(np.max(x_vals), np.max(y_vals), np.max(inv_vals))
plt.plot([sym_min, sym_max], [sym_min, sym_max], color='black', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Function, Inverse')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
