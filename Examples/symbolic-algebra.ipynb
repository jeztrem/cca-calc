##### Use the symbolic algera package, sympy, to find the inverse of $f(x) = 1 - x^2$.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Define the symbols
x, y = sp.symbols('x y')

# Define the function f(x) = 1 - x^2
f = 1 - x**2

# Solve for the inverse: Set y = f(x), then solve for x in terms of y
inverse_eq = sp.Eq(y, f)
inverse_solution = sp.solve(inverse_eq, x)

# Choose one branch of the inverse to ensure it's a proper function (domain restriction)
inverse_func = inverse_solution[0]  # This is x = sqrt(1 - y)

print("Original function: f(x) = 1 - x^2")
print("Inverse function (restricted): f⁻¹(y) =", inverse_func)

# Plotting
x_vals = np.linspace(0, 1, 400)
y_vals = 1 - x_vals**2

y_vals_inv = np.linspace(0, 1, 400)
x_vals_inv = np.sqrt(1 - y_vals_inv)


plt.plot(x_vals, y_vals, label='f(x) = 1 - x²')
plt.plot(y_vals_inv, x_vals_inv, label='Inverse: f⁻¹(y)', linestyle='--')
plt.plot(x_vals, x_vals, color='gray', linestyle=':', label='y = x')  # identity line

plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and Its Inverse')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
