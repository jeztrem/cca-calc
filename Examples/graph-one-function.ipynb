##### import `numpy` for numerical operations.
##### import `matplotlib.pyplot` for plotting.

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Get the set of x values from -a to -a with n values

```python
a = 10
npoints = 250
x = np.linspace(-a, a, npoints)
```

```python
#print(x) #Remove the first hashtag to see the values of x.
```

##### Define the function

```python
y = x**3 + 3*x**2
```

```python
#print(y) #Remove the first hashtag to see the values of y.
```

##### Plot the function on the domain [-a,a]
##### Add a title, labels, grid, labels, and legend

```python
plt.plot(x, y, label=r'$x^3 + 3x^2$', color='purple')
plt.title('Plot of $x^3 + 3x^2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
