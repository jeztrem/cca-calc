##### Examples plots of sin/cos with slope vallues of the tangent lines

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 100)

# Calculate sine and cosine values
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the functions
plt.plot(x, y_sin, label='sin(x)', color='blue')
plt.plot(x, y_cos, label='cos(x)', color='red')
plt.title('Sine and Cosine Functions')
plt.xlabel('x (radians)')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

##### Compute the tangent lines for cos(x), the value of the slope of the tangent lines and compare with the sin function.

```python
# Define x values
x = np.linspace(0, 2 * np.pi, 400)
y_cos = np.cos(x)
y_sin = np.sin(x)

# Tangent points
x_points = np.linspace(0, 2 * np.pi, 10)
y_points = np.cos(x_points)
slopes = -np.sin(x_points)

# Set up subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Top plot: cos(x) with tangent lines
axs[0].plot(x, y_cos, label='cos(x)', color='blue')
for x0, y0, m in zip(x_points, y_points, slopes):
    x_tangent = np.linspace(x0 - 0.5, x0 + 0.5, 10)
    y_tangent = m * (x_tangent - x0) + y0
    axs[0].plot(x_tangent, y_tangent, color='orange', linestyle='--', linewidth=1)
axs[0].set_title('Cos(x) with Tangents')
axs[0].set_xlabel('x (radians)')
axs[0].set_ylabel('y')
axs[0].grid(True)
axs[0].legend()

# Bottom plot: Slopes of tangents and sin(x)
axs[1].plot(x_points, slopes, 'go-', label='Slope = -sin(x)', markersize=6)
axs[1].plot(x, y_sin, label='sin(x)', color='purple', linestyle=':')
axs[1].set_title('Slope of Tangents vs. sin(x)')
axs[1].set_xlabel('x (radians)')
axs[1].set_ylabel('Value')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```

##### Compute the tangent lines for sin(x), the value of the slope of the tangent lines and compare with the cos function.

```python
# Define x values
x = np.linspace(0, 2 * np.pi, 400)
y_sin = np.sin(x)
y_cos = np.cos(x)  # Derivative of sin(x)

# Tangent points
x_points = np.linspace(0, 2 * np.pi, 10)
y_points = np.sin(x_points)
slopes = np.cos(x_points)  # Derivatives

# Set up subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Top plot: sin(x) with tangent lines
axs[0].plot(x, y_sin, label='sin(x)', color='green')
for x0, y0, m in zip(x_points, y_points, slopes):
    x_tangent = np.linspace(x0 - 0.5, x0 + 0.5, 10)
    y_tangent = m * (x_tangent - x0) + y0
    axs[0].plot(x_tangent, y_tangent, color='orange', linestyle='--', linewidth=1)
axs[0].set_title('Sin(x) with Tangents')
axs[0].set_xlabel('x (radians)')
axs[0].set_ylabel('y')
axs[0].grid(True)
axs[0].legend()

# Bottom plot: Slopes of tangents and cos(x)
axs[1].plot(x_points, slopes, 'ro-', label='Slope = cos(x)', markersize=6)
axs[1].plot(x, y_cos, label='cos(x)', color='blue', linestyle=':')
axs[1].set_title('Slope of Tangents vs. cos(x)')
axs[1].set_xlabel('x (radians)')
axs[1].set_ylabel('Value')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```
