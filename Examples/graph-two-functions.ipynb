##### import `numpy` for numerical operations.
##### import `matplotlib.pyplot` for plotting.

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Get the set of x values from -a to -a with n values

```python
a = 10
npoints = 200
x = np.linspace(-a, a, npoints)
```

```python
#print(x) #Remove the first hashtag to see the values of x.
```

##### Define the functions

```python
y1 = x**3 + 3*x**2
y2 = x**2 + 3*x
```

```python
#print(y1) #Remove the first hashtag to see the values of y1.
#print(y2) #Remove the first hashtag to see the values of y2.
```

##### Plot the functions on the domain [-a,a]
##### Add a title, labels, grid, labels, and legend

```python
plt.plot(x, y1, label=r'$x^3 + 3x^2$', color='purple')
plt.plot(x, y2, label=r'$x^2 + 3x$', color='black')
plt.title('Plots')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
