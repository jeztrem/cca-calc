```python
# Problem 31, page 287
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 397)
fx = 3*x**4 + 8*x**3 - 48*x**2
fprime = 12*x**3 + 24*x**2 - 96*x
plt.plot(x, fx, label=r'$3x^4 + 8x^3 - 48x^2$', color='blue')
plt.plot(x, fprime, label=r'$12x^4 + 24x^2 - 96x$', color='red')
plt.title('')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Compute the derivative of fx symbolically using sympy
import sympy as sp
x = sp.symbols('x')
fx_sym = 3*x**4 + 8*x**3 - 48*x**2
fprime_sym = sp.diff(fx_sym, x)
print('Symbolic derivative of fx:')
print(fprime_sym)

# Convert symbolic expressions to numerical functions
fx_num = sp.lambdify(x, fx_sym, modules=['numpy'])
fprime_num = sp.lambdify(x, fprime_sym, modules=['numpy'])

# Generate x values and evaluate functions
x0 = np.linspace(-5, 5, 400)
y_fx = fx_num(x0)
y_fprime = fprime_num(x0)

# Plot the functions
plt.plot(x0, y_fx, label=r'$3x^4 + 8x^3 - 48x^2$', color='blue')
plt.plot(x0, y_fprime, label=r'Derivative', color='red')
plt.title('Function and Derivative')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Compute the derivative of fx symbolically using sympy
import sympy as sp
x = sp.symbols('x')
fx_sym = x**(-2)*sp.log(x)
fprime_sym = sp.diff(fx_sym, x)
print('Symbolic derivative of fx:')
print(fprime_sym)

# Convert symbolic expressions to numerical functions
fx_num = sp.lambdify(x, fx_sym, modules=['numpy'])
fprime_num = sp.lambdify(x, fprime_sym, modules=['numpy'])

# Generate x values and evaluate functions
x0 = np.linspace(1.5, 1.75, 400)
y_fx = fx_num(x0)
y_fprime = fprime_num(x0) 
# Find absolute min and max of fx_sym on [1/2, 4]
interval = [sp.Rational(1,2), 4]
# Find critical points in the interval
critical_points = sp.solve(fprime_sym, x)
critical_points_in_interval = [pt.evalf() for pt in critical_points if pt.is_real and interval[0] <= pt.evalf() <= interval[1]]
# Evaluate at endpoints and critical points
test_points = [interval[0], interval[1]] + critical_points_in_interval
values = [(pt, fx_sym.subs(x, pt).evalf()) for pt in test_points]
min_pt, min_val = min(values, key=lambda t: t[1])
max_pt, max_val = max(values, key=lambda t: t[1])
print(f"\nAbsolute minimum on [1/2, 4]: x = {min_pt}, f(x) = {min_val}")
print(f"Absolute maximum on [1/2, 4]: x = {max_pt}, f(x) = {max_val}")
```
