```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#function on page 127
a = 15
x = np.linspace(-a, a, 400)
fx = (x**2 -1)/(x**2 +1)

# Plot 
plt.plot(x, fx, label=r'$\frac{x^2-1}{x^2 +1}$', color='black')
plt.title(' ')
plt.xlabel('x')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# function on Page 129
a = 10*np.pi
x = np.linspace(-a, a, 400)
fx = np.arctan(x)

# Plot 
plt.plot(x, fx, label=r'$tan^-1$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.grid(True)
plt.legend()
plt.show()
```
