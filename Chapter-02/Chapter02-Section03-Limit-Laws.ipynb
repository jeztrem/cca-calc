```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define x values, compute f(x)
a = 3
x = np.linspace(-a, a, 100)
fx = -x*x
gx = x*x*np.cos(20*np.pi*x)
hx = x*x

# Plot f(x)
plt.plot(x, fx, label=r'${-x^2}$', color='red')
plt.plot(x, gx, label=r'${x^2\cos(20\pi x)}$', color='black')
plt.plot(x, hx, label=r'${x^2}$', color='green')
plt.title('Function')
plt.xlabel('x')
plt.grid(True)
plt.legend()
plt.show()
```
