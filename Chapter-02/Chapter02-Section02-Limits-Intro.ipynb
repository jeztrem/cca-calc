```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define x values, compute f(x)
a = 5
x = np.linspace(-a, a, 400)
y = ((x-1)/(x*x - 1))

# Plot f(x)
plt.plot(x, y, label=r'$\frac{x-1}{x^2 -1}$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define x values, compute f(x)
x = np.linspace(-0.5, 2, 400)
y = ((x-1)/(x*x - 1))

# Plot f(x)
plt.plot(x, y, label=r'$\frac{x-1}{x^2 -1}$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define x values, compute f(x)
x = np.linspace(0.95, 1.05, 100)
y = ((x-1)/(x*x - 1))

# Plot f(x)
plt.plot(x, y, label=r'$\frac{x-1}{x^2 -1}$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define x values, compute f(x)
x = np.linspace(-1.05, 1.05, 100)
y = ((x*x+9)**0.5-3)/(x*x)

# Plot f(x)
plt.plot(x, y, label=r'$\frac{\sqrt{t^2 + 9} - 3}{t^2}$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Define x values, compute f(x)
x = np.linspace(-0.00001, 0.00001, 100)
y = ((x*x+9)**0.5-3)/(x*x)

# Plot f(x)
plt.plot(x, y, label=r'$\frac{\sqrt{t^2 + 9} - 3}{t^2}$', color='black')

plt.title('Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
