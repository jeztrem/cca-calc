```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Problem 21
a = -1
delta = 0.0001
x0 = np.linspace(-3, a, 100)
y0 = x0+3
x1 = np.linspace(a+delta, 3, 100)
y1 = 2**x1

# Plot 
plt.plot(x0, y0, label=r'$x+3$', color='red')
plt.plot(x1, y1, label=r'$2^x$', color='black')
plt.title('Function')
plt.xlabel('x')
plt.grid(True)
plt.legend()
plt.show()
```
