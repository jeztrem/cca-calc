```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, expand, limit

# example of how to obtain analytical solution for problem 23.
t, h = symbols('t h')
f =  2.5*t**2 + 6*t
f_th = 2.5*(t + h)**2 + 6*(t+h)
difference_quotient = (f_th - f) / h
print(difference_quotient)
simplified = expand(difference_quotient)
print (simplified)
derivative = limit(simplified, h, 0)
derivative
```

```python
x, h = symbols('x h')
f =  x**0.5
f_xh = (x + h)**0.5
difference_quotient = (f_xh - f) / h
simplified = expand(difference_quotient)
derivative = limit(simplified, h, 0)
derivative
```

```python
# Define the function and its derivative
def f(x):
    return x**0.5

def f_prime(x):
    return 1/(2*x**0.5)

# Choose a point to draw the tangent line
x0 = 1
y0 = f(x0)
slope = f_prime(x0)

# Define the tangent line: y = m(x - x0) + y0
def tangent_line(x):
    return slope * (x - x0) + y0

# Create x values for plotting
x_vals = np.linspace(0, 5, 400)
y_vals = f(x_vals)
y_prime_vals = f_prime(x_vals)
tangent_vals = tangent_line(x_vals)

# Plot the function and tangent line
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='f(x) = sqrt(x)', color='blue')
plt.plot(x_vals, y_prime_vals, label='f(x) = sqrt(x)', color='green')
plt.plot(x_vals, tangent_vals, label=f'Tangent at x = {x0}', color='red', linestyle='--')
plt.scatter(x0, y0, color='black', zorder=5)
plt.text(x0 + 0.5, y0, f'Point ({x0}, {y0})', fontsize=10)

# Add labels and legend
plt.title('Function and Tangent Line')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x, h = symbols('x h')
f = x**2 + 3*x + 12
f_xh = (x + h)**2 + 3*(x + h) + 12
difference_quotient = (f_xh - f) / h
simplified = expand(difference_quotient)
derivative = limit(simplified, h, 0)
derivative
```

```python
# Define the function and its derivative
def f(x):
    return x**2 + 3*x + 12

def f_prime(x):
    return 2*x + 3

# Choose a point to draw the tangent line
x0 = 1
y0 = f(x0)
slope = f_prime(x0)

# Define the tangent line: y = m(x - x0) + y0
def tangent_line(x):
    return slope * (x - x0) + y0

# Create x values for plotting
x_vals = np.linspace(-5, 5, 400)
y_vals = f(x_vals)
tangent_vals = tangent_line(x_vals)

# Plot the function and tangent line
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='f(x) = x² + 3x + 12', color='blue')
plt.plot(x_vals, tangent_vals, label=f'Tangent at x = {x0}', color='red', linestyle='--')
plt.scatter(x0, y0, color='black', zorder=5)
plt.text(x0 + 0.5, y0, f'Point ({x0}, {y0})', fontsize=10)

# Add labels and legend
plt.title('Function and Tangent Line')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
