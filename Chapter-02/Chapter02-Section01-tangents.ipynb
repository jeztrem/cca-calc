### Chapter 2, Section 1: Tangent and Velocity problems
- The first plot is of a parabola and it's tangent lines.
- The second plot is same using x**3.
- The third plot is same using x**4.
- The values of the slope of the tangents can be computed: If $f(x) = x^n$ then the slope of the tangent is $=nx^{n-1}\$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define x values
a = 5
x = np.linspace(-a, a, 400)
y = x**2

# Points where the tangent lines are computed.
x_points = np.linspace(-a, a, 11)
y_points = x_points**2

# Using the slope formula rise/run or (change in y)/(change = x)
delta = 0.3
slopes = ((x_points + delta)**2 - (x_points - delta)**2)/(2*delta)

# Set up subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 10))

# Plot x*x and tangent lines
axs[0].plot(x, y, label='x*x', color='blue')
for x0, y0, m in zip(x_points, y_points, slopes):
    x_tangent = np.linspace(x0 - delta, x0 + delta, 10)
    y_tangent = m * (x_tangent - x0) + y0
    axs[0].plot(x_tangent, y_tangent, color='red', linestyle=':',  linewidth=1)
axs[0].set_title('x*x and Tangent Lines')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].grid(True)
axs[0].legend()

for x0, y0, m in zip(x_points, y_points, slopes):
    x_tangent = np.linspace(x0 - delta, x0 + delta, 10)
    y_tangent = m * (x_tangent - x0) + y0
    axs[1].plot(x_tangent, y_tangent, color='red', linestyle=':', linewidth=1)
axs[1].set_title('Tangent Lines')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].grid(True)

# Plot slope values
axs[2].plot(x_points, slopes, 'go-', label='Slope = 2x', markersize=6)
axs[2].set_title('Slope of Tangents to x*x')
axs[2].set_xlabel('x')
axs[2].set_ylabel('Slope')
axs[2].grid(True)
axs[2].legend()

plt.tight_layout()
plt.show()
```
